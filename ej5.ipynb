### Oferta y Demanda
- **Demanda:** Se refiere a la cantidad de un bien o servicio que los consumidores están dispuestos a comprar a diferentes precios durante un período determinado. Generalmente, hay una relación inversa entre el precio del bien o servicio y la cantidad demandada: a menor precio, mayor demanda y viceversa.
- **Oferta:** Se refiere a la cantidad de un bien o servicio que los productores están dispuestos a ofrecer a diferentes precios durante un período determinado. Contrario a la demanda, la relación es directa: a mayor precio, mayor cantidad ofrecida.

### Equilibrio
El punto de equilibrio de mercado ocurre cuando la cantidad de un bien que los consumidores están dispuestos a comprar (demanda) iguala la cantidad que los productores están dispuestos a ofrecer (oferta). En este punto se determina el precio de equilibrio y la cantidad de equilibrio del bien o servicio.

### Modelo Simplificado
El modelo simplificado propuesto inicialmente tiene como objetivo ilustrar cómo variaciones en el precio (oferta) afectan la demanda, lo cual es una representación a pequeña escala de los conceptos de oferta y demanda. En este modelo:
- **Precio** influencia la **Demanda.**
- Se asume un **coeficiente de precio** como un valor que representa cómo varía la demanda con variaciones en el precio.
- A través de iteraciones (simuladas con el algoritmo de Euler), se observa cómo, bajo ciertas condiciones y parámetros, cambios en el precio afectan la demanda en el tiempo.

### Como se llegó a ese modelo
Este modelo simple se deriva al enfocar la atención en dos variables críticas, precio y demanda, y asumir una relación lineal entre ellas. Se utiliza un algoritmo numérico (Euler) para simular la evolución del sistema en el tiempo, basándose en una tasa de cambio constante para el precio y observando cómo esto afecta la demanda.

Este tipo de modelos son útiles para entender dinámicas básicas y para la enseñanza de conceptos fundamentales en economía y dinámica de sistemas, aunque en la realidad, los sistemas económicos son mucho más complejos y requieren modelos más detallados y sofisticados para ser comprendidos y predichos de manera adecuada.

```python

def euler_method(demanda, precio, coeficiente_precio, coeficiente_demanda, delta_t, steps, precio_deseado):
    demandas = [demanda]
    precios = [precio]
    
    for _ in range(steps):
        demanda -= coeficiente_precio * precio * delta_t
        precio += (coeficiente_demanda * demanda + (precio_deseado - precio) * coeficiente_precio) * delta_t
        
        demandas.append(demanda)
        precios.append(precio)


        
    return demandas, precios
```

```python
demanda_inicial = 100
precio_inicial = 10
coeficiente_precio = 0.01  # Coeficiente que define cómo el precio afecta la demanda y se auto-corrige.
coeficiente_demanda = 0.01  # Coeficiente que define cómo la demanda afecta el precio.
delta_t = 0.1
steps = 1000
precio_deseado = 10  # Precio al cual se desea llegar mediante auto-corrección.

demandas, precios = euler_method(demanda_inicial, precio_inicial, coeficiente_precio, coeficiente_demanda, delta_t, steps, precio_deseado)
```

```python
import matplotlib.pyplot as plt


plt.plot(demandas, label='Demanda')
plt.plot(precios, label='Precio')
plt.xlabel('Tiempo')
plt.ylabel('Valor')
plt.legend()
plt.show()
```
